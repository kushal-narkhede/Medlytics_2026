# SVM Exercises

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn import metrics
```

Load Pima diabetes dataset

```python
url = "https://raw.githubusercontent.com/BeaverWorksMedlytics2020/Data_Public/master/NotebookExampleData/Week1/diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv(url, names=names)

invalid = ['plas', 'pres', 'skin', 'mass']
for i in invalid:
    data[i].replace(0, np.nan, inplace=True)

data = data.dropna(axis=0).reset_index(drop=True)

data.head()


# YOUR CODE HERE
```

Split data into training, colidation, and test sets with a test size of 25%

```python
# YOUR CODE HERE
X_cols = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']
y_col = 'class'

test_size = 0.25
X_train, X_test, y_train, y_test = train_test_split(data[X_cols], data[y_col], test_size=test_size, random_state=0)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=test_size, random_state=0)
```

Create a model with sklearn's SVC and a high gamma value (perhaps 1.0)

```python
# YOUR CODE HERE
model = SVC(gamma=1, C=1)
model.fit(X_train, y_train)
```

Print accuracy of training and validation data

```python
# YOUR CODE HERE
y_pred_training = model.predict(X_train)
print("Training Accuracy:", accuracy_score(y_train, y_pred_training) * 100, "%")
y_pred_validation = model.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_pred_validation) * 100, "%")
```

Create a model with sklearn's SVC and a low gamma value (perhaps .001)

```python
# YOUR CODE HERE
model2 = SVC(gamma=0.001, C=1)
model2.fit(X_train, y_train)
```

Print accuracy of training and validation data

```python
# YOUR CODE HERE
y_pred_training = model2.predict(X_train)
print("Training Accuracy:", accuracy_score(y_train, y_pred_training) * 100, "%")
y_pred_validation = model2.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_pred_validation) * 100, "%")
```

Explain the accuracy results below. 

When the gamma value was reduced, the training accuracy dropped but the validation accuracy increased. 

Print the confusion matrices for your high and low gamma models (both numeric and visualized)

```python
# YOUR CODE HERE
from sklearn.metrics import confusion_matrix
svm_low = SVC(gamma=0.001, C=1)
svm_low.fit(X_train, y_train)
svm_high = SVC(gamma=1, C=1)
svm_high.fit(X_train, y_train)
y_pred_low = svm_low.predict(X_val)
y_pred_high = svm_high.predict(X_val)
cm_low = confusion_matrix(y_val, y_pred_low)
cm_high = confusion_matrix(y_val, y_pred_high)
print("Confusion Matrix for Low Gamma SVM:\n", cm_low)
print("Confusion Matrix for High Gamma SVM:\n", cm_high)
```
